```python

#The below implementation stroges spare date in an efficient way where time complexity will be (1 â€“ NNZ/mn) NNZ stands
# for Non Zero elements

#The below code create D non overlapping diagonal matix 

import numpy as np
import pandas as pd
import random as rd
from scipy.sparse import csr_matrix
n=int(input("Enter the value of n for rows and column:"))
d=int(input("Enter the value for block size diagonal matrix:"))
#The dimensions of Matrix A is given by number of block diagonal matrix and the rows * column
f=n*d
#Temp variables used 
temp_a=0
temp_b=0
A=np.zeros(shape=(f,f))
for i in range (0, f, d):
    for j in range (0, f, d):
        
        while((temp_a < f) & (temp_b < f)):
            A[temp_a][temp_b]=rd.randint(1,9)
            temp_a+=1
            temp_b+=1
            
            
        A[i][j]=rd.randint(1,9)
        temp_a=i
        temp_b=j
        
print(A)
"""
Storing Sparse matrix alogorithm
Step 1: Set M to number of rows in MATRIX
Step 2: Set N to number of columns in MATRIX
Step 3: I = 0, NNZ = 0. Declare A, JA, and IA. 
        Set IA[0] to 0
Step 4: for I = 0 ... N-1
Step 5: for J = 0 ... N-1
Step 5: If MATRIX [I][J] is not zero
           Add MATRIX[I][J] to A
           Add J to JA
           NNZ = NNZ + 1
        [End of IF]
Step 6: [ End of J loop ]
        Add NNZ to IA
        [ End of I loop ]
Step 7: Print vectors A, IA, JA
Step 8: END
"""
NNZ=0
A1_value=[]
JA_col=[]
IA_rowptr=[0]

for i in range (0, len(A)):
    for j in range (0, len(A[0])):
        if(A[i][j]!=0):
            A1_value.append(A[i][j])
            JA_col.append(j)
            NNZ=NNZ+1
    
    IA_rowptr.append(NNZ)
    
print(A1_value)
print(JA_col)
print(IA_rowptr)
```
